```python
##FunciÃ³n para importar excel o csv
def importarDatos(ruta, extension,sepa=None):
  if extension == 'xlsx':
    return pd.read_excel(ruta)
  elif extension == 'csv':
    return pd.read_csv(ruta, sep=sepa)
  else:
    print('Formato no soportado')

datoexcel = importarDatos('/content/drive/MyDrive/DatosCurso/CovidDiagnosis.xlsx','xlsx')
datoexcel

datoscsv = importarDatos('/content/drive/MyDrive/DatosCurso/Datos.csv', 'csv',';')
datoscsv
```
