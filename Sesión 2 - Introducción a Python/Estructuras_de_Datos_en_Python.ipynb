### ***1. Introducción a Listas*** ⛪

```python
listas = [1,True,3,2.3,"hola", False,"hola"]
type(listas)
```

```python
### recorrido de la lista para conocer el tipo de dato
for i in listas:
    print(type(i))

for i in listas:
  print(f"el tipo de dato de {i} es {type(i)}")
```

```python
##Propiedades de las listas
#indexación simple
print(listas[0])
#Indexación por rangos de indice
print(listas[1:4])
#Indexación desde cero hasta indice 2
print(listas[:3])
#Indexación desde indice 2 hasta el final
print(listas[2:])
#Indexación por rangos y saltos
print(listas[1:6:2])
#Agregar elementos a la lista
#listas.append(23)
#print(listas)
##conteo de elementos
listas.count("hola")
#inserción de datos en una posición en particular
listas.insert(2,100)
print(listas)
##Eliminar datos de las listas
listas.remove(23)
```

```python
#Eliminar por posición o por el último
listas.pop(2)
listas
```

```python
##unir listas
lista1 = [1,2,3]
lista2 = [4,5,6]
lista2.extend(lista1)
lista2

##operaciones matematicas con listas
lista2 + lista1
```

```python
##Creación de dataframes con listas
import pandas as pd
df = pd.DataFrame(lista2)
df
```

### ***2. Introducción a Diccionarios*** ⛪

```python
#ejemplo de diccionario
d = {'edad':39, 'nombres':'Julian Quimbayo', 'Curso':'Data Science'}
type(d)
```

```python
print(d['nombres'])
##impresion completa del diccionario
print(f"El nombre y apellido del profesor es: {d['nombres']}, su edad es {d['edad']} y el curso que dicta es {d['Curso']}")
```

```python
###Solo las claves o solo los valores
claves = list(d.keys())
valores = list(d.values())
print(claves)
print(valores)
```

```python
#Extraer los datos completos claves y valor en forma de lista
datog = list(d.items())
datog
```

```python
##Agregar valores al diccionario
d['pais'] = 'Perú'
d
```

```python
##sobreescribir valores del diccionario
d['pais'] = 'Colombia'
d
```

```python
###Eliminar datos con valores o claves
del d['pais']
d
```

```python
d.pop('edad')
d
```

```python
##Construir dataframes con diccionario
import pandas as pd
df = pd.DataFrame(d,index=[0])
df
```

### ***5. Funciones*** ⛪

```python
##forma de crear una función
def hola(nombre, edad):
  print(f"Hola Mundo {nombre}: y su edad es: {edad}")

hola("Julian Quimbayo", 39)

hola("Alexander Morales", 24)
```

```python
###funcion para sumar datos de una lista
def sumar(lista):
  suma=0
  for i in lista:
    suma += i
  return suma

sumar([1,2,3,4,5])
```
