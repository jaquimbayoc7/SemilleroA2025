### **5. Normalizaci贸n**

```python
##https://www.linkedin.com/pulse/standardization-normalization-regularization-nikita-sharma-jauvf#:~:text=Standardization%20and%20Normalization%20are%20data,doesn%27t%20affect%20the%20data.

from sklearn.preprocessing import MinMaxScaler
def normaData(df):
    valores = df.values
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaler = scaler.fit(valores)

    pd.DataFrame(np.vstack((scaler.data_min_, scaler.data_max_)),
             index=['Min', 'Max'],
             columns=df.columns)

    normalizados = scaler.transform(valores)
    df_norm = pd.DataFrame(normalizados,
                       index=df.index,
                       columns=df.columns)
    return df_norm

estnorma = normaData(est)

estnorma.describe().T
```

### **6. Test de Normalidad**

```python
##D'agostino, shapiro-wilk, jarque-bera, kolmogorov - smirnov , anderson-darling.
from scipy import stats
def testShapiroWilk(df):
        valoresP = []
        concepto = []
        variable = []
        for column in df:
            k2, p_value = stats.shapiro(df[column].values)
            valoresP.append(p_value)
            variable.append(column)
            if (p_value < 0.05):
                concepto.append('No es una variable Normal')
            else:
                concepto.append('Es una variable Normal')
        dfShapiro = pd.DataFrame(
            {'Variable': variable, 'Valores P': valoresP, 'Concepto': concepto})
        return dfShapiro

testShapiroWilk(est)
```

### **10. Heteroscedasticidad y Multicolinealidad**

La heteroscedasticidad es cuando la varianza de los errores en un modelo de regresi贸n no es constante. Vamos a explicarlo de manera simple.

La heteroscedasticidad puede causar:

Estimaciones menos precisas
Intervalos de confianza incorrectos
Pruebas de hip贸tesis menos confiables

Test de Breusch-Pagan:
H: Homoscedasticidad
H: Heteroscedasticidad
Si p-valor < 0.05: Rechazar H (hay heteroscedasticidad)

Soluciones si se Detecta
Transformaci贸n de Variables:
Logar铆tmica
Ra铆z cuadrada
Box-Cox
Regresi贸n Ponderada (WLS):
Asigna diferentes pesos a las observaciones
Errores Est谩ndar Robustos:
Corrige los errores est谩ndar sin cambiar los coeficientes

```python
import statsmodels.api as sm

# Definir las variables independientes y la variable dependiente
X = estnorma[['age', 'traveltime', 'famrel',
       'freetime', 'health', 'absences','G2']]
y = estnorma['G3']

# A帽adir una constante a las variables independientes para el intercepto
X_with_const = sm.add_constant(X)

# Ajustar el modelo de regresi贸n
model = sm.OLS(y, X_with_const).fit(cov_type='HC3')

# Obtener los valores ajustados (predicciones) y los residuos
fitted_vals = model.fittedvalues
residuals = model.resid

# Mostrar el resumen del modelo (opcional)
print(model.summary())
```

```python
# Gr谩fico de residuos vs predicciones
plt.figure(figsize=(8, 6))
plt.scatter(fitted_vals, residuals, color='blue', alpha=0.6)
plt.axhline(y=0, color='red', linestyle='--')
plt.title('Gr谩fico de residuos vs valores ajustados')
plt.xlabel('Valores ajustados (predicciones)')
plt.ylabel('Residuos')
plt.show()
```

```python
from statsmodels.stats.diagnostic import het_breuschpagan

# Prueba de Breusch-Pagan para homocedasticidad
test_results = het_breuschpagan(residuals, X_with_const)

# El resultado devuelve cuatro valores: Lagrange multiplier statistic, p-value, f-value, f p-value
labels = ['Lagrange Multiplier Statistic', 'p-value', 'f-value', 'f p-value']
print(dict(zip(labels, test_results)))
```

1. El valor p te indica si puedes rechazar la hip贸tesis nula de homocedasticidad.
2. Si el valor p es menor que 0.05, indica que hay heterocedasticidad.
3. Si el valor p es mayor que 0.05, no hay evidencia suficiente para rechazar la homocedasticidad.

La multicolinealidad es un fen贸meno estad铆stico que ocurre cuando existe una alta correlaci贸n entre variables independientes en un modelo de regresi贸n. Esto significa que dos o m谩s variables predictoras est谩n fuertemente relacionadas entre s铆.

Problemas que causa:
Estimaciones inestables de los coeficientes de regresi贸n
Errores est谩ndar inflados
Dificultad para interpretar la importancia individual de cada variable
Reducci贸n en la precisi贸n de las predicciones

Factor de Inflaci贸n de la Varianza (VIF)
VIF > 5-10 indica problemas de multicolinealidad

Soluciones
Eliminar variables altamente correlacionadas
Combinar variables relacionadas
An谩lisis de componentes principales (PCA)
Regularizaci贸n (Ridge, Lasso)
Aumentar el tama帽o de la muestra

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
# Calcular el VIF para cada variable independiente
vif_data = pd.DataFrame()
vif_data["Variable"] = X_with_const.columns

# El VIF se calcula usando el variance_inflation_factor de statsmodels
vif_data["VIF"] = [variance_inflation_factor(X_with_const.values, i) for i in range(X_with_const.shape[1])]

# Mostrar los resultados de VIF
print(vif_data)
```

Interpretaci贸n de los resultados de VIF:

    1. Un VIF de 1 significa que no hay multicolinealidad entre esa variable y las dem谩s.
    2. Un VIF entre 1 y 5 indica niveles aceptables de multicolinealidad.
    3. Un VIF mayor a 5 indica una multicolinealidad alta, lo que puede ser problem谩tico para el modelo. (sqrt , logn)

### **10. Introducci贸n a Regresi贸n Lineal Simple**

```python
##60% train 40% test - 70% train 30% test - 80% train 20% -90% train 10% test
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

##Seleccionar las variables
##Variable independiente
X = estnorma[['G2']]
##variable a predecir o dependiente
y = estnorma['G3']

##Particionar los datos -siempre
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.40, random_state=42)

##crear el modelo
modelouno = LinearRegression()
modelouno.fit(X_train, y_train)

##Predicciones
y_pred = modelouno.predict(X_test)

##Metricas
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

##Calculo del R^2
r2 = round(r2_score(y_test, y_pred),3)

##Imprimir los resultados
print('Mean Squared Error (MSE)', mse)
print('Root Mean Squared Error (RMSE)', rmse)
print('R^2 Score:', r2)
```

```python
##Grafica del modelo
X_range = pd.DataFrame(np.linspace(X.min(), X.max(),100))
y_pred = modelouno.predict(X_range)

plt.figure(figsize=(10,6))
plt.scatter(X_train, y_train, color='blue', label='Datos Entrenamiento')
plt.scatter(X_test, y_test, color='green', label='Datos Prueba')
plt.plot(X_range, y_pred, color='red', label='Modelo Predicci贸n')
plt.title('Modelo Uno de regresi贸n lineal g2 Vs g3')
plt.xlabel('Variable independiente X')
plt.ylabel('Variable dependiente Y')
plt.legend()
```
